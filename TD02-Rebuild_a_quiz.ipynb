# Create parent class that represents the format or structure of our QCM

```python
class Font:
    """ font theme for printing/display ergonomic text"""
    PURPLE = '\033[95m'
    CYAN = '\033[96m'
    DARKCYAN = '\033[36m'
    BLUE = '\033[94m'
    GREEN = '\033[92m'
    YELLOW = '\033[93m'
    RED = '\033[91m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'
    END = '\033[0m'


class QCM:
    """ PARENT: format des Question à Choix Multiples. """
    def __init__(self, question:str, choix_multiples:list, bonne_reponse:str):
        self.question = question
        self.choix_multiples = choix_multiples
        self.bonne_reponse = bonne_reponse
        
    def __str__(self, num=None):
        # print(f"{num}. Question: {self.question}:")
        # for index, choix in zip("abcdefghi", self.choix_multiples):
        #     print(f"\t{index} {choix}")
        # print(f"==> Answer: {self.bonne_reponse}\n")
        choix = [f"{index}. {choix}" for index, choix in zip("abcdefghi", self.choix_multiples)]
        str_question = Font.BOLD + f"- Question {num}) {self.question}\n\t" + Font.END
        str_choix = Font.RED + f"{"\n\t".join(choix)}" + Font.END
        str_reponse = Font.GREEN + f"\n==> Answer: {self.bonne_reponse}\n" + Font.END
        # template = str_question + str_choix + str_reponse
        template = str_question + str_choix
        return template
    
    def __repr__(self):
        return f"object = QCM(question={self.question}, choix_multiples={self.choix_multiples}, bonne_reponse={self.bonne_reponse})\n"


# Quiz database format
QUIZZ_DATA = [
    QCM(
        "Combien de fois la France a gagné la coupe du monde ? ", # Question
        ["1", "5", "3", "2", "0"], # Choix Multiples [a, b, c, d, e, ...]
        "d" # Reponse
    ),
    QCM(
        "Quand a été fondé Apple ? ",
        ["1985", "2026", "1976", "1945"],
        "c",
    ),
    QCM(
        "Qui a fondé SpaceX ? ",
        ["David Guetta", "Elon Musk", "Bill Gates", "Louis de Funès"],
        "b"
    ),
    QCM(
        "Quelle est la capitale de la France ?",
        ["Paris", "Berlin", "Madrid", "Londres"],
        "a"
    ),
    QCM(
        "Combien font 2 + 3 ?",
        ["4", "5", "6", "7", "0"],
        "b"
    ),
    QCM(
        "Qui est le vainqueur de l'EU Champion league 2026 ?",
        ["Barça", "PSG", "Real Madrid", "Donald Trump", "Stéphanie de Monaco"],
        "b"
    ),
    QCM(
        "Completer la suite de ces nombres: 1 - 5 - 13 - ?? - 37",
        ["0", "20", "50", "23"],
        "d"
    ),
    QCM(
        "Completer la suite de ces nombres: 9 - 25 - ?? - 81",
        ["50", "64", "49", "72"],
        "c"
    ),
    QCM(
        "Combien de pays en Afrique ?",
        ["54", "100", "24", "12", "32"],
        "a"
    ),
    QCM(
        "En Histoire & Géographie, le Monde est divisé en combien de continents ?",
        ["4", "3", "7", "12", "6", "5"],
        "c" # EN Histoire/Geographie, 7 grands continents : l'Amérique du Nord, l'Amérique du Sud, l'Europe, l'Asie, l'Afrique, l'Océanie, et le plus petit de tous, l'Antarctique. L'Europe et l'Asie sont collées l'une à l'autre, alors certains pensent qu'il n'y a que 6 continents au total. 
    )
]

# Checking of QCM object
for num, question in enumerate(QUIZZ_DATA):
    print(question.__str__(num+1))
```

# Create child class that represents the launcher of our QCM

```python
class Launch_Quizz(QCM):
    """ CHILD : Run quizz & calculate score. """
    def __init__(self, QUIZ_database:list, question:str="", choix_multiples:list=[], bonne_reponse:str=""):
        super().__init__(question, choix_multiples, bonne_reponse)
        self.score = 0
        self.total_questions = 0
        self.QUIZ_database = QUIZ_database
        self.num_chance = 0

    def evaluate_score(self, correct_answer=False):
        self.total_questions += 1
        if correct_answer:
            self.score += 1

    def score_in_percentage(self):
        if self.total_questions == 0:
            return 0
        return (self.score / self.total_questions) * 100

    def display_final_score(self):
        pourcentage = self.score_in_percentage()
        print("\n+++++++++++++++++++++++++++++++++++++++++++++++++++++++++")
        print(f"The quizz is finished!!!,\n\tYour final Score is : {self.score}/{self.total_questions} ==> {pourcentage}%")
        if pourcentage >= 80.0: # Seuil de passage de l'examen  > 80% sinon echec!!!
            print(f"\n{Font.GREEN}******** CONGRATULATIONS: You are the quiz WINNER!!! ********{Font.END}")
        else:
            print(f"\n{Font.RED}******** LOSER: Oh no, you lost the quiz... BAD!!! ********{Font.END}")
        print("+++++++++++++++++++++++++++++++++++++++++++++++++++++++++\n")
        
    def run_and_manage_responses(self,):
        for num, qcm in enumerate(self.QUIZ_database):
            print(qcm.__str__(num+1))
            response = qcm.bonne_reponse
            user_answer = ""
            self.num_chance = 3
            correct_answer = False
            while (user_answer != response) and (self.num_chance != 0):
                # user_answer = input("Enter your answer :").strip().lower()
                user_answer = input("Enter your answer :").lower()
                print(f"Your answer : {user_answer}")
                correct_answer = user_answer == response.lower()
                if correct_answer:
                    print("==> Good job! This is the right answer!\n")
                    break
                else:
                    self.num_chance -= 1
                    print(f"==> Sorry, you have {self.num_chance} chances left!!!")
            self.evaluate_score(correct_answer)
            print("")


if __name__ == "__main__":
    # Run the quiz
    test1 = Launch_Quizz(QUIZZ_DATA)
    test1.run_and_manage_responses()
    test1.display_final_score()
```
